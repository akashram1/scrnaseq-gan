## Cell Type Distributions

```python
count1 = 0
for c in cell_types1:
    cell_type_count = len(labels1[labels1['Characteristics[cell type]'] == c])
    count1 += cell_type_count
    print(c, str(cell_type_count))
print("Total usable cells = ", str(count1))
```

```python
## Labels(Note: Gamma and PP cells are same.)
## PP  = Pancreatic Polypetide cell
cell_type_dict_1 = {'alpha cell': 0, 
                    'beta cell': 1,
                    'delta cell': 2,
                    'gamma cell': 3}


cell_type_dict_2 = {'cell subtype: alpha': 0, 
                    'cell subtype: beta': 1,
                    'cell subtype: delta': 2,
                    'cell subtype: PP': 3}
```

## Adding Phenotype cols
### Dataset 1

```python
mat1['cell_type'] = pd.Series(dtype=int, index=mat1.index)
mat1['disease_condition'] =  pd.Series(dtype=int, index=mat1.index)
```

```python
for cell_id in mat1.index:
    cell_type = labels1[labels1['Source Name'] == cell_id]['Characteristics[cell type]'].values[0]
    dis_cond =  labels1[labels1['Source Name'] == cell_id]['Characteristics[disease]'].values[0]
    
    if str(cell_type) not in cell_type_dict_1.keys():
        mat1.loc[cell_id]['cell_type'] = -1
    else:
        mat1.loc[cell_id]['cell_type'] = cell_type_dict_1[cell_type]
    
    if str(dis_cond).rstrip() not in dis_cond_dict_1.keys():
        mat1.loc[cell_id]['disease_condition'] = -1
    else:
        mat1.loc[cell_id]['disease_condition'] = dis_cond_dict_1[dis_cond]
```
